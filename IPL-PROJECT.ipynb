```python
maxwins = df['match_winner'].value_counts()
sns.barplot(y = maxwins.index, x = maxwins .values)
plt.title('Most Match Winers')
```

```python
sns.countplot(x = df['toss_decision'])
plt.title('Toss Decision')
```

```python
POM = df['player_of_the_match'].value_counts().head(10)
```

```python
sns.barplot(x = POM.values,y = POM.index)
```

```python
Max_Runs = df.groupby('top_scorer')['highscore'].sum()
Max_Runs
```

```python
Max_Runs = df.groupby('top_scorer')['highscore'].sum().sort_values(ascending = False)
Max_Runs
```

```python
High = df.groupby('top_scorer')['highscore'].sum().sort_values(ascending = False).head(2)
High
```

```python
High.plot(kind = 'bar') #barh to make the graph Horizontal
```
