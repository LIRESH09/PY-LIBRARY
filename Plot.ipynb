```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X = np.linspace(0,5,11)
Y = X**2
Y
```

```python
plt.plot(X,Y)
plt.title("X-Y Relation")
plt.xlabel("X-Values")
plt.ylabel("Y-Values")
```

SUB PLOT

```python
plt.subplot(2,2,1) # (2,2) are the Row and Columns
plt.plot(X,Y)
plt.subplot(2,2,2) # (1,2,3,4) are the Loaction where the graph should be Show
plt.plot(X,Y)
plt.subplot(2,2,3)
plt.plot(X**2,Y**2)
plt.subplot(2,2,4)
```

OBJECT ORIENTED WAY


Matplotlib allows the aspect ratio, DPI and figure size to be specified when the Figure object is created. You can use the figsize and dpi keyword arguments.

figsize is a tuple of the width and height of the figure in inches dpi is the dots-per-inch (pixel per inch).

```python
fig = plt.figure(figsize=(8,6),dpi=120)                    #White Frame
axis1 = fig.add_axes([.1,.1,0.6,0.8]) #(0.1)is the Reduce the Size from Left
axis1.plot(X,Y)                       #(0.1)Reduce the Size from Right
axis2 = fig.add_axes([.4,.3,.5,.4])   #(0.6) is the Width
axis2.plot(X,Y)                       # (0.8) is the Height 
```
