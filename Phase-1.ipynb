```python
import numpy as np
import time
```

Creation of Array from List  in Python

```python
arr_1 = np.array([1,4,3])
print("Array-1",arr_1)

arr_2 = np.array([[1,4,3],[3,1,4]])
print("Array-2",arr_2)
```

Multiplication

```python
list = (1,2,3)
print("List Multiplication :",list)

arr = np.array([1,2,3]) #Element wise Multiplication
print("Array Multiplication :",arr*2)
```

Efficiency of Time in List (or) Array

```python
#In List
start = time.time()
py_list = [1 * 2 for i in range(1000000)]
print("\n List Operation Time :", time.time() - start)
#In Numpy
start = time.time()
py_arr = np.arange(1000000) * 2
print("\n Numpy Operation Time :", time.time() - start)
```

Creating Different Types of Array's in Numpy

```python
Zero = np.zeros((4,5))
print("Zero's Array :\n",Zero)

Ones = np.ones((4,4))
print("\nOnes's Array :\n",Ones)

Full = np.full((4,4),3) #To print a Specific Number
print("\nFull's Array :\n",Full)

Random = np.random.random((2,3))
print("\nRandom's Array :\n",Random)
```

Vector - Matrices - Tensor

```python
Vector = np.array([1,2,3,4,5])
print("Vector's: ",Vector)

Matrix = np.array([[1,2,3,4],
                    [5,6,7,8]])
print("\nMatrices's: ",Matrix)

Tensor = np.array([[1,2,3],[4,5,6], #This a Multi-Dimensional Array 
                   [7,8,9],[1,2,3]]) 
print("\nTensor:",Tensor)
```

Array Properties

```python
arr = np.array([[1,2,3],
                [4,5,6]])
print("Shape: ",arr.shape) #It Gives you Rows and Columns
print("Dimension: ",arr.ndim) #It Gives you Dimension
print("Size: ",arr.size) #It Gives you Total number of Elements
print("Data Type: ",arr.dtype) #It Gives you Type of DataType
```

Array Reshaping

```python
arr = np.arange(16)
print("Original Array:",arr)

reshaped = arr.reshape([4,4])
print("\nReshaped Array:\n",reshaped)

Flattened = reshaped.flatten()
print("\nFlatten Array:\n",Flattened)
 
#Gives you (Returns Original View,Instead of Copy) 
Raveled = reshaped.ravel()
print("\nRaveled Array:\n",Raveled)

Transpose = reshaped.transpose()
print("\nTransposed Array:\n",Transpose)
```
