```python
Coffee = pd.read_csv('coffee.csv')
Coffee
```

Accessing Data With Pandas 

```python
# Coffee.head()             # To access the First 5 Rows of a DataSet only integer's

# Coffee.tail(10)           # To access the Last 5 Rows of a DataSet --- .tail(10) is used to get last 10 Rows 

# Coffee.sample(10)         # To get Random Rows of a DataSet 

# Coffee.loc["Money"]           # Loc is used to modify rows and columns in DF
```

```python
Coffee.sort_values(["money"],ascending=False)
```

Filtering Data

```python
Coffee.loc[Coffee['money'] > 20 , ['coffee_name','money' ]]  # Without Condition

Coffee.loc[(Coffee['money'] > 20 ) & (Coffee['coffee_name'] == 'Latte' )]   # With Condition 

Coffee[Coffee['coffee_name'].isin(["Espresso"])]  

```
