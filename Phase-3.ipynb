Advanced Operations In Python

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Data Structure : [restaurant_id, 2021, 2022, 2023, 2024]
sales_data = np.array([
    [1, 300000 , 350000 , 450000 , 500000], #Shadab
    [2, 250000 , 300000 , 280000 , 350000], #Shah Ghouse
    [3, 200000 , 250000 , 300000 , 400000], #Paradise
    [4, 350000 , 280000 , 350000 , 400000], #Bawarchi 
    [5, 260000 , 320000 , 380000 , 400000]  #Mehfil
])

print("-------ZOMATO SALES ANALYSIS-------")
print("\n Sales Data Sahpe :",sales_data.shape)
print("\n To print Data of First 3 Sales :\n",sales_data[:3])
```

```python
#Total Sales per Year 
print("Sum of the Sales of Year :\n",np.sum(sales_data,axis = 0))
print("\nYearly Total :\n",np.sum(sales_data[:,1:],axis=1))

#Min Sales
print("\nMin Sales in Year :\n",np.min(sales_data[:,1:],axis = 1))

#Max Sales
print("\nMax Sales in Year :\n",np.max(sales_data[:,1:],axis=1))

#Average Sales
print("\nAverage Sales in Year :\n",np.mean(sales_data[:,1:],axis=1))

#Cumulative Sales
Cumulative = np.cumsum(sales_data[:,1:],axis=1)
print("\nCumulative Sales in Year :\n",Cumulative)

#Matplotlib
plt.figure(figsize=(6,4))
plt.plot(np.mean(Cumulative,axis=0))
plt.title("Avg Cumulative sales across Restaurant ")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

Vector's in Numpy

```python
vector1 = np.array([1,2,3,4,5,6])
vector2 = np.array([7,8,9,10,11,12])
#Vector Addition
print("Vector Addition :",vector1 + vector2)
#Vector Subtraction
print("Vector Suntraction :",vector1 - vector2)
#Vector Multiplication
print("Vector Multiplication :",vector1 * vector2)
#Vector Dot Product
print("Vector Dot Product :",np.dot(vector1 , vector2))
#Vectorization 
Restaurant_types = np.array(["Shadab","Shah Ghouse","Paradise","Bwarachi","Mehfil"]) 
Vectorization = np.vectorize(str.upper) #Goes to Every Single Element 
print("Vectorized Upper:",Vectorization(Restaurant_types))
```
